## From Biological Intelligence to Artificial Intelligence

### 1 Introduction

Previous development of D-M model for polygenic inheritance produced a model (see Binomial Genomics) that can account for independent gene configurations. But this did not account for interactions between genes and gene-environment interactions. It was proposed a hierarchical tree model (Bayesian or HMM) would be required to generate a complete model. Here we consider that an alternative to the conventional hierarchical graph systems could be explored by a deeper understanding of how biological neural networks function. Drawing on this understanding we develop fundamental principles for a new machine learning paradigm from the ground up. This would involve significant change in thinking from basic (Boolean) logic of computer science and machine learning to incorporating neuromorphic features in affordable large scale integrated circuits such as FPGAs. This will be in line with the basic principles and algorithms developed in this set of experiments.

The aim is to find strategies for using discrete, binarized recurrent networks, such that D-M distribution replaces Bayesian networks and bit-wise recurrent processes replace HM like methods. By simplifying the process in this way, it is hoped to build a framework that is congruent with biology and could reduce costs and improve efficiency. The realization of the Dirichlet (Alpha)-Multinomial algorithm was achieved deductively based on the basic assumptions of Mendelian inheritance and Hardy-Weinberg equation. Increasing generalizations were derived to obtain the final model using Dirichlet (alpha) multinomial distribution for estimating polygenic gene configurations. 

### 2 Description of the Book

Here we start with the higher brain networks, in which recurrent feedback processes occur (from single neurons to large networks) with capacity to generate complex information processing including mental rotations and navigational tasks from physical to social. It is the intention of this study to generate and test hypotheses that can develop a model of recurrent biological networks that can account for information processing at all levels of the biopsychosocial spectrum:

1)	**Individual neurons** (feedback mechanisms and logic processing)
2)	**Cortical Modules** and feedback processes that lead to mental rotations
3)	**Physical navigation** and feedback processes important to navigation including processing of landmarks and emotions
4)	Proposal of **emotions** taking the form of **abstract landmarks** for abstract navigation mixed with physical navigation
5)	Need to consider both the sphere (for rotation) and the plane (for navigation), with the former having a direct impact on the latter.
6)	Build the mathematical case for sphere rotation and navigation on the **complex plane** being models for **feedback processes** on the neural tree.
7)	Join this feedback model using **binarized logi**c as the edges and **D-M distribution** at each level as nodes to build the hybrid machine learning model that generalizes processes in the biopsychosocial spectrum from genes to social navigation.

### 4 Developing AI Code

The final stage is to develop the machine learning algorithm based on the empirical and deductive knowledge above:

1)	Generate the Machine Learning Algorithm using the mathematical models as a basis (this is important, as the **biology** informs the **mathematical models** that in turn informs the **machine learning model**. This provides a logical connection between biological networks and their simulation in a machine learning paradigm).
2)	Build a hybrid model using **D-M model** and **XOR XNOR recursion**:
  
**Stage One-** Use pure rotational dynamics to test the emergence of membrane dynamics (because membrane dynamics is independent of gene configurations and neural networks, this would be an independent test of a model derived from assumptions about polygenic function and interactions of a network).
  
**Stage Two-** Test the impact of the initial model using pure XOR-XNOR recursion (rather than sphere rotation) with D-M hybrid on membrane dynamics.
  
**Stage Three-** Further testing of the model on simple image recognition and interaction of a Chromosome with epigenetic factors.
  
**Stage Four-** Foundation for development of a scale invariant canonical machine learning model that could act as the Kernel for a broad machine learning paradigm in the health field. We will call this model Life Emulating Orchestrated Network (Life emulating, given that the deductive models have biological processes as their foundation and Orchestrated due the nature of the information processing that has a graph like structure reminiscent of a symphony and in honour of the Penrose-Hameroff model OrchOR).
