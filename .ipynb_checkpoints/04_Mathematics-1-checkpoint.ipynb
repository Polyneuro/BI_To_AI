## Building the basic mathematical building Blocks

### A)	Geometric Algebra

Geometric Algebra (GA) Basics: 
  
‚Ä¢ GA extends vector algebra by introducing a geometric product, which combines the dot (inner) and wedge (outer) products. The geometric product of two vectors $(\mathbf{a} \ and \ \mathbf{b} )$ is:
  
$\mathbf{a}\mathbf{b} = \mathbf{a} \cdot \mathbf{b} + \mathbf{a} \wedge \mathbf{b}. $

‚Ä¢ The wedge product $\mathbf{a} \wedge \mathbf{b}$ represents an oriented plane (a bivector), generalizing to higher-dimensional objects (trivectors, etc.) in higher dimensions.

Multivectors: 
‚Ä¢ A multivector is a sum of elements of different grades (scalars, vectors, bivectors, etc.). For example, in 3D GA, a general multivector $ M $ might be written as:
$M = \alpha + \mathbf{v} + \mathbf{B} + \mathbf{T} + \beta I,$

where
$ alpha $ is a scalar, $ \mathbf{v} $ is a vector, $ \mathbf{B} $ is a bivector, $ \mathbf{T} $ is a trivector, and $ I $ is the pseudoscalar (highest-grade element).

### Bivectors, trivectors and higher elements in GA: 

Bivectors: 
‚Ä¢ In GA, the concept of a bivector (e.g., of two vectors) can be elegantly expressed using the geometric product or by leveraging reflections. 
  
For two unit vectors $\mathbf{a}$ and $\mathbf{b}$, the bivector is often proportional to $\mathbf{a} + \mathbf{b}$ (analogous to the diagonal of the parallelogram they span).
  
‚Ä¢ This extends to hyperplane bivectors in higher dimensions, where GA provides compact formulas for reflections and rotations (e.g., using versors).
  



  <img src="images/vectors_parallelepiped.png" style="display:block; margin:auto" width="500"/>

The number of elements of a multivector is determined by the number of vector elements (determining the dimension of the base-space).
  
For two dimensions there is one scalar, two vectors and on pseudoscalar, which is. bivector ($2^{2}$ elements).
For three dimensions, there is one scalar, three vectors, three bivectors and one trivector which is the pseudoscalar ($2^{3}$ elements).
  
Similarly for four dimensions there are $2^{4}$ elements (4 vectors and 4 trivectors and 6 bivectors along with the scalar and pseudoscalar).
  
Thus, the elements of multivectors in geometric Algebra are configured according to the binomial distribution (an interesting observation given the previous derivation of binomial aspect of gene configurations).

### Connection between the Riemann sphere and Bivector-planes


Returning to the equation relating the complex plane to angles on the sphere surface, it is possible to relate these to both i = ‚àö-1 on the complex plane and $ B = e_{1}e_{2} $ on a plane spanned by a bivector:

$ \cot\left(\frac{\phi}{2}\right) e^{-i\theta} \quad \text{becomes} \quad \left(\frac{\cos(\phi/2)}{\sin(\phi/2)}\right) e^{-B\theta} $

Interpretation: 
  
‚Ä¢ On the Riemann sphere, a point can be parameterized by spherical coordinates $ (\phi, \theta) $:

$ \phi $: Polar angle (from the north pole).
  
$ \theta$: Azimuthal angle (in the equatorial plane).
  
‚Ä¢ The left side uses complex numbers $( i = \sqrt{-1} )$, while the right side replaces $ i $ with a bivector $ B $ (e.g., $ B = e_1e_2 $) for the GA formulation.
  
‚Ä¢ The cotangent term $ \cot(\phi/2) $ scales the magnitude, while ($ e^{-B\theta}$ ) rotates in the bivector plane.
Geometric Insight: 
  
‚Ä¢ This resembles the stereographic projection from the Riemann sphere to the complex plane, but now expressed in GA terms.


$e^{-B\theta/2} a e^{B\theta/2}$ = $e^{-B\theta}a$ = a $e^{B\theta}$

Interpretation:
  
‚Ä¢ Sandwich Product: The transformation $e^{-B\theta/2} a e^{B\theta/2}$ rotates the vector $ a $ by angle $\theta$ in the plane defined by $ B $ (this is how rotors work in GA).
  
‚Ä¢ Equality to $e^{-B\theta}a$: This holds only if $ a $ lies entirely in the plane of $ B $. If $a$ has components orthogonal to $B$, those components remain unchanged (this could be interpreted as $ \theta$ changing (rotation on the plane) while $ \phi $ remains the same).
  
‚Ä¢ Commutativity Note: The last equality $e^{-B\theta}a = a e^{B\theta}$ is true only if $a$ is in the plane of $B$ (since $a$ anti-commutes with $B$ in that case). Otherwise, it‚Äôs more nuanced.
  
Geometric Insight:
  
‚Ä¢ This is the GA version of rotating a vector by $\theta$ in the plane B. The sandwich product ensures rotation is grade-preserving (a vector stays a vector).

**M√∂bius Transformations in GA:**
M√∂bius transformations on the Riemann sphere can be expressed in GA using rotors and reflections. 
  
For example:
  
‚Ä¢ Translation/Rotation/Scaling: Represented by even-grade multivectors (rotors).
  
‚Ä¢ Inversion/Reflection: Represented by odd-grade multivectors (e.g., vectors).

A general M√∂bius map
$$f(z) = \frac{az+b}{cz+d}$$
  
in GA might involve:
  
$f(a) = (a \cdot \sigma_1)^{-1} (a \cdot \sigma_2)$
  
where, 
( $\sigma_1, \sigma_2 $) are spinors (rotors) derived from the coefficients $( a, b, c, d )$.

It should be noted that a M√∂bius transformation on the Riemann sphere is a rotation that involves a change in both polar and azimuthal angles. This is a subset of all possible M√∂bius transformations. Here, when referring to M√∂bius transformation, we mean a rotation of the Riemann sphere.

**Primary vs Secondary (Social Emotions)**
As indicated previously, primitive sense of self-boundary functions emerge very early in evolution of multi-cellular organisms and the structures involved can be phylogenetically followed from the lower brainstem all the way to emergence of Telencephalon and the cignulate gyrus in mammals.
  
The primitive emotions in fish can be considered largely retricted to **figh-or flight** reactions and **vigilance**. The structures equivalent to amygdala and hippocampus, providing the **emotional** reactions and the **spatial** context beigin to emerge and covers the forebrain. It may be assumed that forerunners of emotions such as **joy** and **sadness** may involve feeding and resting when food is not available. The emotional circuit involves mainly connection to the effector circuit via the hypothalamus and input from the PAG. The differentiation between aggression and sexual activity may not be highly developed. Later, land animals develop the striatum, necessary for assuming posture against gravity. With the emergence of the ventral striatum, amygdala-ventral striatum connection result in **avoidance**, **approach** behaviour. In these lower regions, states such as **anger-fear**, **joy-sadness** may assume a more recognisable form by humans. This becomes more evident when mammals emerge with the **cingulate gyrus**, facilitating social emotions and navigation. It is of interest that structures such as the **insula**, lies on the outside of the cortex in mammals such as mouse and rat (inner aspects of the lateral sulcus or **Sylvian fissure**). Emergence of this structure may have facilitated the more explicit experience of disgust (surprise and disgust being opposite poles of reactions to an initial stmulus) and surprise. Surprise is essentially a primitive emotion associated with early attention for potential threat. In mammals, with the advnt of the higher cortical structures all the primitive feelings become more nuanced (e.g. surprise-disgust axis would be influenced by interest or aversion to new relationships, as well as food or other stimuli) and can be recognises as **primary emotions**.
  
It is postulatd here that, with mammalian evolution, the central emotional circuits (amygdala, ventral striatum and spatial context from hippocampus) have **modulatory activity** both from the **higher centers** (frontal systems including the cingulate) and continued stimulation from **lower centres** such as the Periaquiductal Grey (**PAG**) and the mono-aminergic projections from the RAS to both the ventral striatum and hypothalamus. This dual modulation of the function of the central limbic structures probably produce the primary emotions we can recognise in mammmals such as humans via recurrent circuits.
  
Given their early association with landmarks and spatial context, emotions could be recognised as emerging as **abstract landmarks** when navigating physical space and later social space. Here we generate a model of emotions as landmarks in a **physical** as well as **social** landscape permitting rapid decisions including **simulation** before action. Further, we use **antipodal relations** between emotions. It is proposed that this permits smooth and if necessary rapid transitions in emotional stance whether one is far or close to home, or among close humans or strangers.
