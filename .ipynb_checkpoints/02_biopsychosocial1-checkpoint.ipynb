## Biopsychosocial spectrum as a system of recurrent feedback and homeostatic equilibrium - gene interaction at intra and intercellular level

As we have explored the theoretical basis for polygenic gene configurations already and outlined the likely gene candidates for polygenic inheritance in common disorders such as major depression and chronic pain, here we explore the idea of recurrence from ground up, starting with genomics.

### 1 Polygenic influence on biological phenotypes

Science of genomics has progressed significantly since it departed from the Central Dogma of the theory of evolution, to recognize the importance of interaction between the genes and environment; that gene expression is dynamic, including the potential for acquired changes to be passed on to the germ line after several generations.

While this does not dismiss the well-recognized mechanisms in the transcription and translation of the genetic code on the DNA-complement in to proteins, regulatory regions in the DNA have been increasingly recognized. Such regulation is so ubiquitous, that the diagram of biological networks of interaction below displays gene expression as a relatively small part of the overall biological network.

Nevertheless, change of gene expression in which it finds a new equilibrium can have profound effect, especially in thought, mood and behavior. We explore the most likely biological concomitants that mediate such changes.

### 2 Honing in on biologically effective interactions on the Neural Network- The amines

Monoamine (serotonin, noradrenalin, dopamine, acetylcholine), glutamate and GABA (external) receptors, the metabolic and catabolic enzymes of these amines, can influence a wide range of neural networks, from single neurons to cortical modules and large networks. They act in a broad manner (i.e. large actions on cortical modules) that could be inferred from external recordings when coupled with their gene expression and epigenetic over lay. For example, release of Damage Associated Molecular Patterns (DAMPs) due to stress hormones can influence a cascade of intracellular (including nuclear) and extra-cellular events that lead to the Blood Brain Barrier (BBB) becoming more permeable to peripheral immune cells. These cells enter the CNS and can be marked by the epigenetic environment of the brain. The peripheral cells can then be analyzed for their genome to consider both inherited and epigenetic (acquired) vulnerabilities. Bringing together brain imaging patterns and peripheral genomic patterns can theoretically be used in a machine learning paradigm to infer models of network activity of an individual‚Äôs biopsychosocial network.

Below is the example of a metabotropic receptor that may be influenced by ligands such as serotonin. They commonly have 7 protein subunits that make up the complex of receptor ion channel and metabolic enzyme that can respond to the binding of the receptor including activation of the ion channel. The second messenger cascade that result from the enzyme activity can in turn affect cellular activity including (indirectly) gene expression. Thus, even at the immediate vicinity of a single cell there exists a circular cascade that mediates gene-environment interaction.

### 4 Neurotrophic factors, second messengers, inflammasome (intra and intercellular systems)

Neurons and microglia release the DAMP, HMGB1 in the presence of cortisol that impacts on Toll like receptor 4 (TLR4). This causes a cascade in which NFùúÖB is produced and enters the nucleus and acts on the promoter of the NLRP3. This primes the inflammasome path to be sensitive to future stress or inflammation. The inflammasome is responsible for cleavage of pro-caspase that then cleaves pro-IL1 to IL1. IL1 in turn can affect the production of BDNF (a neurotrophic factor that can dampen inflammation and promote synaptic connections) and turnover of 5HT and Noradrenalin, by affecting gene expression of elements in the pathways of monoamine synthesis and degradation.

### 5 Membrane channels (boundary function of single neurons)

<img src="images/Ion_channel.png" style="display:block; margin:auto" width="500"/>

Source: https://commons.wikimedia.org/wiki/File:Ion_channel.png#Licensing

Neural membrane function is influenced largely through interaction between ion-channels, specifically Na+, K+ and Cl-. The ion channels involved in the cascade of events leading to spike trains are voltage gated (their function is affected by the level of voltage at the membrane). This means that the flow of ions through the semi-permeable membrane is affected by the charge difference across it which ion-flow affects this charge difference in turn.

Membrane channel production can be influenced by their gene expression although less work is done in this area. Ion channel can be affected by environmental changes and can change their level of sensitivity and clustering on the membrane by dynamically moving within the membrane. While the expression of components that affect the function of ion-channels are expressed via the genetic make-up and could be affected by epigenetic mechanisms, they are likely less important in global behaviour than monoamine and inflammatory function. In this series, this is nevertheless important to consider because membrane function is ultimately in the pathway between genes and neuronal function, thus ultimately in the cascade that leads to the overall function of neural networks that leads to emergence of social navigation, as bridge towards the social aspects of the biopsychosocial cascade.
