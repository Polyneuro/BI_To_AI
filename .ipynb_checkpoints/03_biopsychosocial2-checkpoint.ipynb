## Biopsychosocial Spectrum - Recurrence, from neurons to networks and navigation

So far, we have discussed recurrent circuits from the lowest levels between genes and their microenvironment to processes that spans the global function of the brain and body (e.g. monamine-inflammasome-stress diathesis). It is also important to close this set of loops by considering such circuits that can occur within a single neuron up to large neural networks. The aim of this exercise is to demonstrate that recurrent feedback circuits are ubiquitous. Further, given the demonstration of a binomial type tree in polygenic expression (see binomial genomics repository), the fact that neurons and neural networks naturally have a treelike structure indicates that recurrence and network graphs is a natural outcome of interactions of different systems levels, in the biopsychosocial spectrum. Here, we will focus on the recurrent networks and some of the phenotypic characteristics of their function.

### 1 Single Neurons and recurrent feedback

First, the single neuron is characterized on its dendritic arbor by a feed forward of signals but also feedback of signals called back-propagation that facilitates logical computation, specifically XOR. Most pyramidal neurons in the cortex have input from all six amines described.

### 2 Larger Neural Networks and recurrent feedback

Next level of complexity are local networks. The canonical network in this regard is the cortical module that is characterized by specific inputs and outputs and a recurrent loop within it to amplify signals.

<img src="images/Cortical-Module.png" style="display:block; margin:auto" width="500"/>

Interaction between cortical modules have been associated with a phenomenon of mental rotations. Such rotations are necessary ensure perceptual modules are invariant under rotation (say of the body or head). This is achieved with recurrence between modules including facilitatory and inhibitory (feedback) paths. In addition to the experience of rotation, the same circuits are also characterized as sensory and memory components. The latter is necessary for comparison purposes. These inter-module circuits occur either between adjacent or distant modules representing recurrence in the processing of information that is in the same modality or wider recurrence up to considering sensory versus action-oriented processing. It is important to note that feed forward projections are mostly from the superficial pyramidal layers to other cortical areas, while feedback from other cortical areas occur at the apical tuft of deep (layer 5 pyramidal neurons). It is the opposite in relation to lower brainstaem structures (input to layer 5 and feedback from layer 2-3). 

The most complex circuits involve that of the Default Mode, Central Executive and Salience Networks. Building up to these networks is the evolution of navigation in the external world, evolved at the level of the hippocampus and extending to engage increasingly remote centers, resulting in increasingly abstract notions of “navigation”.

### 3 Hippocampus, Amygdala, Cingulate Gyrus – From physical navigation to emotions as abstract landmarks

The basic building blocks of navigation requires mapping of the environment in relation to self. In the figures below, we see the parts of the hippocampus (CA1, CA2, CA3, dentate gyrus and the subiculum) and associated cortical areas in the para-hippocampal gyrus (PHCG).

Place cells are found in CA1, CA3 and the dentate gyrus while grid cells are found in the PHCG. There are also head-direction cells (entorhinal cortex, presubiculum, parts of the PHCG), landmark vector cells (CA1) and boundary vector cells (subiculum). Together, these cells serve the function of mapping the environment, setting boundaries and navigation in a defined space.

<img src="images/Hippocampus.png" style="display:block; margin:auto" width="500"/>

Firstly, the hippocampus could be considered as the location to which spatially relevant information is funneled and acts as a memory processing unit for mapping the external space (independent of other cues such as smell input directly to the amygdala that interacts with the hippocampus, and distant visual cues from other cortical areas). Head direction cells are cued by distant landmarks but are independent of them. These cells appear to use the azimuthal angle for their computations. Head direction and place cells are intimately connected.

Interestingly, the position of the embodied self is important for driving the place cells and their ability remap a region independent of distant landmarks. Most probably **Precuneus** (see human brain below) provides input to place cells (via the PHCG) to fulfill this function. The body movement and sense of time appear to be used by place cells to remap an area. External information is necessary to correct any drifting of the egocentric map from the external world (allocentric). Essentially, hippocampus integrates path integration (summation of inferred distances from velocity and time sense) with cues from external landmarks. Further, the environment requires a horizon or boundary against which movements and the map is formulated.

**Conceptual Construct of Evolution of Brain strutures of Emotions and Navogation:**  **Mouse**(top) and **Human** (below) Hippocampus in relation to the rest of the brain can be seen in the figures above. In the **fish**, the areas equivalent to the amygdala and hippocampus essentailly take up almost all of the Telenecephalon (forebrain). As can be seen, by the time of evolution of the mammals these regions are small compared to the neo-cortex. In fish these structures connect directly to areas such **hypothalamus** for primitive motivational behaviour. In land animals the **amygdala** and **hippocampus** connect to the **ventral striatum** as well with emergence of **operant conditioning**. In mammals another superordinate set of structures from the **frontal lobes** begin to control these ciscuits through **corticostriate loops**. For example, the **anterior cingulate** inhibits the **amgdala-ventral striatum** path, thus inhibiting habit forming behaviour and providing modulation by **secondary social emotions**. More primitive sense of self (forming the embodied boundary) emerges in structures that are in the **mesencephalon** (mid-brain). These are the structures of the **Reticular Activating System (RAS)**, the **Raphe** (origin of seortonin or 5HT neurons), **Locus coeruleus** (orgin of Noradrenalin cells), **Substantia Nogra** (Dopamine projection to the striatum) and **Ventral Tegmentum** (Dopamine projection to the ventral striaum, important for formations of habits such as **avoidance** and **addiction**). Close to and dorsal to these structures is an important structure **Periaqueductal Gray (PAG)** which is a central structure that connects **sensory cues** (thalamus, colliculus) to **motor circuits** (RAS, hypothalamus). Even more primitive than this in evolution of motivated behaviour is the **parabrachial complex** (dorsolateral pons) which receives input from more primitive sensory systems in the **rhomboncephalon** or medulla (**nucleus tractus solitarius**). So, during evolution, increasing overlay of later paths over earlier paths appear to have ocurred. Given that evolution is driven primarily by survival, it could be hypothesised that a **sensory-motor** boundary defining **self vs environment** has always been a factor in emergence of functions of increasing complexity. Here we construct a model of **navigation in space** as an important expression of this fundamental **motivation** to survive.

### 4 Reciprocal connections in navigation and emotions

There are reciprocal connections between the **hippocampus** and the **amygdala** and topographically interdigitating outputs from both to the ventral striatum, which also receives input from the **medial prefrontal cortex**. **Amygdala** receives input from the **orbito-frontal cortex**. The **ventral striatum** is also part of the **cingulate thalamic cortico-striate** loop. While the **hippocampus** does not receive reciprocal input from the **ventral striatum**, there is a circuit that leads back from the **cingulate cortex** to the **hippocampus**.

There are several models of how the hippocampus engages in spatial mapping and navigation. Principal models are attractor model and oscillation models. It is likely that an appropriate model of mapping and navigation requires conjoint information processing of sensory (outer world) and egocentric and internal stimuli including simulation of motor activity. Recurrent circuits are again ubiquitous in the mapping and navigation circuits.

The interaction with the amygdala and the ventral striatum involves association of place fields with approach avoidance behaviour, wrongly labelled as classical conditioning, whereas it is more likely operant conditioning. In this context it is also of interest to note that the hippocampus is important in temporal coding using place cells as well and thus facilitating anticipatory behaviour as occur in operant conditioning. (It is more likely that classical conditioning is mediated via amygdala hypothalamic connections. 

Thus, the evolution of spatial navigation was probably evolved in conjunction using association with primary emotions for survival. Even in fish, the medial and lateral pallia engage in fear related associations and spatial associations respectively. This was probably the evolutionary predecessors of the amygdala and the hippocampus respectively.

Even though, the hippocampus is necessary for development of maps, path integration (involving summation of direction and distance) does not require the hippocampus once it is memorized. However, path integration will acquire errors in time and requires updating using **border vector cells** and **landmark vector cells** periodically. While models of angle and distance from landmarks have been proposed as a mechanism, there are no satisfactory models that simplifies how circuits involving the hippocampus may achieve this task.

In this study we propose a model that adds to previous models by involving not only the **azimuthal angle** from the horizon but also the **polar angle** that might together provide a method of navigational and mapping computation. While the model is mathematical, it is demonstrated that this is equivalent simply to a cascade of recurrent circuits using particular complementary logic (feed forward facilitatory and feedback inhibitory). 

Such recurrent circuits using a neural graph like structure is demonstrated to not only be sufficient to explain navigation and mapping but all levels of the neural network and can be generalized to **gene networks** and **social networks** as well.
Here we propose that emotions evolved initially associated with physical land marks but have acquired the status of landmarks themselves to facilitate rapid decisions about motor actions. This can be extended to secondary emotions and social navigation associated with the evolution of the **cingulate gyrus** in mammals. Given the relationship of **recurrent circuits** to navigation and mental rotations and even more abstract concepts such as using **emotions as land marks** in abstract navigation tasks, it is appropriate to explore a mathematical model of mapping involving rotation.
