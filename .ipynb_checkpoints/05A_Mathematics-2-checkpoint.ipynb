## Building the basic mathematical building Blocks - Antipodality and XOR-XNOR planes

### A)Antipodality as a mathematical construct 

**Antipodal points** on the **Riemann sphere** relate to the points on the plane with relation, $z_{2} = -1/\bar{z}_1$. This established relation indicates that the points on the plane are symmetrically represented in relation to the unit sphere. A point on the plane outside the unit circle is inverted in its magnitude and it lies on the same orientation with respect the line between the origin and the point outside but negated, so that it is reflected on the plane orthogonal to the line. 

However, the relationship of two orthogonal complex planes in relation to antipodal points on the sphere has not been previously considered (as far as we are aware). Using Geometric Algebra, because rotation of ùúã/2 on ‚ÑÇ2 space, is equivalent to a rotation of ùúã on the sphere (S2), and considering rotation on the sphere as a M√∂bius transformation, the relationship of $z_{2}$ on the orthogonal plane (‚ÑÇ2) to $z_{1} on the first plane (‚ÑÇ1), would be

$z_{2} = -1/\bar{z}_1$ , if the same point on the sphere is projected to both planes. 

This related to the fact that antipodal points on the sphere project to co-ordinates with the same orientation on the plane but symmetrically reflected on the unit circle (algebraic proof of this is below).


**1) any two orthogonal planes in $\mathbb{R}^4$ project to antipodally related points on $\mathbb{S}^2$ $\iff z_{2} = -1/\bar{z}_1$.**

This is essentially a geometric property of **orthogonal planes in $ \mathbb{R}^4$** under **stereographic projection**. Here's the cleanest way to derive it.  


‚∏ª


**1. Setup: Orthogonal Planes in $\mathbb{R}^4$**

Consider two **2D planes** $\Pi_1$ and $\Pi_2$ in $\mathbb{R}^4$, spanning subspaces such that if:
‚Ä¢ $\Pi_1 = \text{span}(\mathbf{u}, \mathbf{v})$
‚Ä¢ $\Pi_2 = \text{span}(\mathbf{w}, \mathbf{x})$


then all basis vectors satisfy:
  
$\mathbf{u} \cdot \mathbf{w} = \mathbf{u} \cdot \mathbf{x} = \mathbf{v} \cdot \mathbf{w} = \mathbf{v} \cdot \mathbf{x} = 0$.



This means **every vector in $\Pi_1$ is orthogonal to every vector in $\Pi_2$**.


‚∏ª


**2. Stereographic Projection from $\mathbb{R}^4$ to $\mathbb{S}^2$**

We use the generalized stereographic projection from $\mathbb{R}^4$ to $\mathbb{S}^2 \subset \mathbb{R}^3$.  


For a **point $\mathbf{p} = (x_1, x_2, x_3, x_4) \in \Pi_1$**:
  
$\text{Projection}(\mathbf{p}) = \left( \frac{2x_1}{\|\mathbf{p}\|^2 + 1}, \frac{2x_2}{\|\mathbf{p}\|^2 + 1}, \frac{\|\mathbf{p}\|^2 - 1}{\|\mathbf{p}\|^2 + 1} \right)$,
  
where $ \|\mathbf{p}\|^2 = x_1^2 + x_2^2 + x_3^2 + x_4^2 $.  


**Key Observation**: Since $ \Pi_1$ and  $\Pi_2$ are orthogonal, we can choose coordinates **without loss of generality** such that: 
  
‚Ä¢ $ \Pi_1 $ is given by $ (x_1, x_2, 0, 0) $,
  
‚Ä¢ $ \Pi_2 $ is given by $ (0, 0, x_3, x_4) $.


This is possible because **any two orthogonal planes** in $\mathbb{R}^4$ can be rotated into this form.


‚∏ª


**3. Projecting Points from $\Pi_1$ and $\Pi_2$**

**Projection of a point $\mathbf{p}_1 = (a, b, 0, 0) \in \Pi_1$**:

  
$(X_1, Y_1, Z_1) = \left( \frac{2a}{a^2 + b^2 + 1}, \frac{2b}{a^2 + b^2 + 1}, \frac{a^2 + b^2 - 1}{a^2 + b^2 + 1} \right)$



**Projection of a point $\mathbf{p}_2 = (0, 0, c, d) \in \Pi_2$**:

  
$(X_2, Y_2, Z_2) = \left( \frac{2c}{c^2 + d^2 + 1}, \frac{2d}{c^2 + d^2 + 1}, \frac{c^2 + d^2 - 1}{c^2 + d^2 + 1} \right)$



‚∏ª


**4. Relationship between Projections**

Now, suppose **$\mathbf{p}_1 \in \Pi_1$ and $\mathbf{p}_2 \in \Pi_2$ are related such that their projections are antipodal**.  


We want to find the condition such that:
  
$(X_2, Y_2, Z_2) = (-X_1, -Y_1, -Z_1)$.



**Matching Coordinate Conditions**
1. **X and Y coordinates**:
  
$\frac{2c}{c^2 + d^2 + 1} = -\frac{2a}{a^2 + b^2 + 1}$,
  

$\frac{2d}{c^2 + d^2 + 1} = -\frac{2b}{a^2 + b^2 + 1}$.
  
These imply:
  
$\frac{c}{d} = \frac{a}{b}$,
  
meaning $ (c, d) = \lambda(-a, -b)$ for some $ \lambda $.

2. **Z coordinate**:
  
$\frac{c^2 + d^2 - 1}{c^2 + d^2 + 1} = -\left( \frac{a^2 + b^2 - 1}{a^2 + b^2 + 1} \right)$.
  
Substituting $(c, d) = \lambda(-a, -b)$, we get:
  
$\frac{\lambda^2(a^2 + b^2) - 1}{\lambda^2(a^2 + b^2) + 1} = -\left( \frac{a^2 + b^2 - 1}{a^2 + b^2 + 1} \right)$.
  
Solving this, we find:
  
$\lambda = \frac{1}{\sqrt{a^2 + b^2}}$.
  
Thus, the only solution for antipodality is:
  
$(c, d) = \left( - \frac{a}{a^2 + b^2}, - \frac{b}{a^2 + b^2} \right)$.



‚∏ª


**5. Conclusion: Antipodal Correspondence**

This proves that:
1. For any **$(a, b, 0, 0) \in \Pi_1$**, its **antipodal point** on $\Pi_2$ is:
  
$(0, 0, c, d) = \left( 0, 0, - \frac{a}{a^2 + b^2}, - \frac{b}{a^2 + b^2} \right)$.
  
2. Under stereographic projection, these map to $ (X_1, Y_1, Z_1)$ and $ (-X_1, -Y_1, -Z_1)$, respectively.


**Thus, *any* two orthogonal planes in $\mathbb{R}^4$ project to antipodally related points on $\mathbb{S}^2$ $\iff z_{2} = -1/\bar{z}_1$.**


‚∏ª


**6. Final Note**

This is very general‚Äîit does not require complex analysis. The key steps are:
1. Align the orthogonal planes in $\mathbb{R}^4$ with coordinates $ (x_1, x_2, 0, 0)$ and $ (0, 0, x_3, x_4) $.
2. Project points from each plane and solve for antipodality.
3. The solution reduces to **inversion (negation + rescaling)**.


We use this result and those below to assert why XOR/XNOR (orthogonal in $\mathbb{R}^4$) appear as **negated (antipodal) on $\mathbb{S}^2$**. (see section (C) )

**Summary with extended working out of the intermediate steps**

If these coordinates are projected from antipodal points on the sphere, then

(X2, Y2, Z2) = (-X1, -Y1, -Z1)

=>	2c/(c2+d2+1) = -2a/(a2+b2+1) =>   2c(a2+b2+1) = -2a(c2+d2+1)
  
=>	2d/(c2+d2+1) = -2b/(a2+b2+1) =>   2d(a2+b2+1) = -2b(c2+d2+1)

Let,  (a2+b2+1) = M and   (c2+d2+1) = N

Then 2cM = -2aN 
And   2dM = -2bN =>  N = -d/b M  
  
=> 2cM = -2a (-d/b) M => c = a(d/b) and c/d = a/b

This means that (c,d) = ùúÜ (a, b)

Using the Z coordinate

(c2+d2-1) /(c2+d2+1) = - [ (a2+b2-1) /(a2+b2+1)]


[ ùúÜ2(a2+b2)-1) /ùúÜ2(a2+b2)+1] = - [ (a2+b2-1) /(a2+b2+1)]

	Let (a2+b2) = p2

=>	2ùúÜ2 p4 = 2 => ùúÜ = 1/p2 = 1/(a2+b2)
  
This is a general proof of antipodal points on the sphere projecting to the coordinates on the orthogonal planes, then the coordinates would related by the relation:
  
(c,d) = (-a/(a2+b2), -b/(a2+b2))

This is based on real coordinates but this can be generalized to orthogonal complex planes. In this case,
  
(c, d) = c + id = - [(a/(a2+b2) + ib/(a2+b2) ] = -[a+ib/(a2+b2)] = -[a+ib/ (a+ib) (a-ib)] = -1/a-ib

If a + ib = z then c+id = -1/zÃÖ
  
Thus, antipodal points on the sphere project to coordinates on the orthogonal planes which have the relationship w = -1/zÃÖ.
Further, these coordinates have the opposite orientation (angle ùúÉ), differing in scaling or magnitude (the magnitude is a reflection on the unit circle and negated due to negation). This makes the transformations involved anti-conformal. This is important when considering relationship of these rotations to logical recursion of orthogonal logic.


**2) antipodally related points on $\mathbb{S}^2$ $\iff z_{2} = -1/\bar{z}_1$ relationship means that any sphere point projects to points on the orthogonal planes such that $z_{2} = -1/\bar{z}_1$.**

Given the relationship of antipodality on the sphere that requires the relationship between orthogonal planes, C1 and C2 to be z2 = -1/zÃÖ1, we can next consider the situation in which the antipodal point is negated and brought to same point. In other words, both projections from the two planes project to the same spherical point, (X,Y,Z).

Then, the projection to C1 can be considered X+iY/1-Z (formula for stereographic projection)
  
Given that the Riemann sphere is a unit sphere (X2+Y2+Z2 = 1),
  
$z_{1}$ = X+iY/1-Z = [(X+iY)*(X-iY)]/[(1-Z)*(X-iY)]
  
= X2+Y2/[(1-Z)*(X-iY)] = 1-Z2/[(1-Z)*(X-iY)] 
  
= [(1+Z)*(1-Z)]/[(1-Z)*(X-iY)] = **1+Z/(X-iY)** 

It is possible demonstrate that this geometric projection when symmetrically inverted and negated provides the previous antipodal relationship between the orthogonal planes.

Using the relationship
    
$z_{2} = -1/\bar{z}_1$ produces $z_{2}$ = -1/[1+Z/(X+iY)]
   
= -(X+iY)/1+Z = (-X) + i(-Y)/ 1-(-Z)

Thus, starting with the same sphere point, the relationship z2 = -1/zÃÖ1, causes a sign inversion, the equivalent of sphere antipodality that was proved to be compatible with orthogonality of the complex planes if z2 = -1/zÃÖ1 is satisfied.

The projection from the same sphere point, with the condition z2 = -1/zÃÖ1 satisfied is able to project to two dual points on the orthogonal planes (symmetrically inverted and negated orientation).
Thus, the same point on the sphere represents the relationship between the orthogonal planes that are symmetric inversion and negation, in the same way that antipodal points on the plane projects to points on a complex plane. One plane represents the antipodal version of the other, projected from the same sphere point.


**3)	Similar argument could be applied to the situation in considering how antipodal points on the sphere project to the orthogonal complex planes:**

For the antipodal point $(-X, -Y, -Z)$ on the sphere, its projection to C‚ÇÅ is:


$$w_1 = \frac{-X + i(-Y)}{1 - (-Z)} = \frac{-X - iY}{1 + Z}$$  (here we use $w_{1}$ to denote projection $‚ÑÇ_{1}$ from the antipodal point).


And for the original point $(X, Y, Z)$, its projection to C‚ÇÇ is:


$$z_2 = -\frac{1}{\overline{z_1}} = -\frac{1}{\frac{X - iY}{1 - Z}}$$


Let's calculate this step by step:


$$z_2 = -\frac{1 - Z}{X - iY}$$


To simplify this, we will multiply both numerator and denominator by the complex conjugate of the denominator:


$$z_2 = -\frac{(1 - Z)(X + iY)}{(X - iY)(X + iY)} = -\frac{(1 - Z)(X + iY)}{X^2 + Y^2}$$


For a point on the unit sphere, $X^2 + Y^2 + Z^2 = 1$, so $X^2 + Y^2 = 1 - Z^2$:


$$z_2 = -\frac{(1 - Z)(X + iY)}{1 - Z^2} = -\frac{(1 - Z)(X + iY)}{(1 - Z)(1 + Z)} = -[\frac{X + iY}{1 + Z}]$$


So the formula for $z_2$ redcues to:


$$z_2 = -[\frac{X + iY}{1 + Z}] = \frac{-X - iY}{1 + Z}$$


Comparing with $w_1$:


$$w_1 = \frac{-X - iY}{1 + Z}$$


We see that $w_1 = z_2$, which confirms the duality relationship! (On orthogonal planes antipodal points always project to the dual planes with the relationship $z_{2} = -1/\bar{z}_1$ and similarly with the same sphere point; when the projecting with respect to the orientation of the anipodal point the relationship reverses $w_{1} = -1/\bar{w}_2$).

**Geometric Interpretation of the Duality using the poles as the antipodal points.** 
  
Let us say that the setreo graphic projection to $‚ÑÇ_{1}$ uses the South pole and the projection to $‚ÑÇ_{2}$ uses the North pole.

This duality arises because:

1. **C‚ÇÅ projection** uses the South Pole $(0,0,-1)$ as the projection point
2. **C‚ÇÇ projection** uses the North Pole $(0,0,1)$ as the projection point

When the these antipodal points project to the two planes, they project to the same co-ordinates such that $z_1$ = $z_2$.

Let us retrun to the previous observation that the same sphere point projects to the planes such that $z_{2} = -1/\bar{z}_1$
Then the antipodal point will project similarly, where $w_{1} = -1/\bar{w}_2$. Thinking about antipodal points as N and S will clarify this (just as reversal of the projection point at infinity reverses the relationship on one plane, it would do the same for the relationship between tow orthogonal planes).

When we project from opposite poles:
  
‚Ä¢ A point $P$ on the sphere projects to $z_1$ on C‚ÇÅ and $z_2$ on C‚ÇÇ
  
‚Ä¢ The antipodal point $-P$ projects to $w_1 = z_2$ on C‚ÇÅ and $w_2 = z_1$ on C‚ÇÇ


This creates the perfect symmetry or duality that can be visulaised using python code (see below).

### B)‚ÑÇ1-‚ÑÇ2 Antipodal Demo - Below is python code that generates an animation of the mathematics dicribed above

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
import matplotlib.colors as mcolors

# Add these at the top of your code (before creating the figure to avoid animation size limit)
plt.rcParams['animation.embed_limit'] = 35  # Increase to 35MB (default is 20MB)
plt.rcParams['figure.dpi'] = 72  # Reduce from default 100 dpi

# Set up the figure with 4 subplots (2x2 grid)
fig = plt.figure(figsize=(18, 12))
ax_sphere = fig.add_subplot(221, projection='3d')  # Riemann sphere (3D)
ax_C1 = fig.add_subplot(222)                      # Projection onto ‚ÑÇ1 (stereographic)
ax_C2 = fig.add_subplot(223)                      # Projection onto ‚ÑÇ2 (antipodal)
ax_4D = fig.add_subplot(224, projection='3d')     # 3D visualization of orthogonal planes
plt.suptitle("Riemann Sphere: Alternating N-S ‚Üî E-W Rotations with Fixed Points and Their Antipodes")

# Unit sphere coordinates
u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi, 50)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones_like(u), np.cos(v))

# Fixed point (unchanging in 3D space)
fixed_point = np.array([0.7, 0.3, 0.5])
fixed_point /= np.linalg.norm(fixed_point)

# Projection functions
def project_C1(p):  # South Pole ‚Üí Plane (stereographic)
    return (p[0] + 1j * p[1]) / (1 - p[2] + 1e-10)

def project_C2(p):  # North Pole ‚Üí Plane (antipodal)
    # Changed to enforce the antipodal relationship
    z1 = project_C1(p)
    return -1 / np.conj(z1 + 1e-10)  # Adding small epsilon to avoid division by zero

# Add this function to visualize relationships between points
def add_relationship_visualization(ax_C1, ax_C2, z1, z2, z1_antipodal, z2_antipodal):
    """
    Add visual aids to show the relationship between C1 and C2 projections.
    """
    # Calculate the direct antipodal point on C1 plane
    direct_antipodal = -1 / np.conj(z1)

    # Calculate magnitudes
    mag_z1 = np.abs(z1)
    mag_z2 = np.abs(z2)

    # 1. Draw line from origin through z1 (extended)
    if mag_z1 > 0:
        extension_factor = 4.0 / mag_z1  # Extend to edge of plot
        ax_C1.plot([0, z1.real * extension_factor], [0, z1.imag * extension_factor],
                  'b--', alpha=0.5, linewidth=1)

    # 2. Mark the direct antipodal point on C1 plane
    ax_C1.scatter(direct_antipodal.real, direct_antipodal.imag, color='purple', s=30,
                 marker='x', label='Direct -1/zÃÑ')

    # Draw line from origin to direct antipodal
    ax_C1.plot([0, direct_antipodal.real], [0, direct_antipodal.imag],
              'purple', linestyle='--', alpha=0.7, linewidth=1)

    # Connect z1 to z1_antipodal with a dashed line
    ax_C1.plot([z1.real, z1_antipodal.real], [z1.imag, z1_antipodal.imag],
              'c--', alpha=0.7, linewidth=1)

    # Connect z2 to z2_antipodal with a dashed line
    ax_C2.plot([z2.real, z2_antipodal.real], [z2.imag, z2_antipodal.imag],
              'g--', alpha=0.7, linewidth=1)

    # Add text annotations for magnitudes
    ax_C1.annotate(f"|z‚ÇÅ| = {mag_z1:.2f}", xy=(z1.real, z1.imag),
                  xytext=(z1.real + 0.2, z1.imag + 0.2),
                  color='blue', fontsize=8)

    ax_C2.annotate(f"|z‚ÇÇ| = {mag_z2:.2f}", xy=(z2.real, z2.imag),
                  xytext=(z2.real + 0.2, z2.imag + 0.2),
                  color='green', fontsize=8)

    # Add product annotation
    ax_C1.annotate(f"|z‚ÇÅ|¬∑|z‚ÇÇ| = {mag_z1 * mag_z2:.2f}",
                  xy=(0.05, 0.05), xycoords='axes fraction',
                  color='black', fontsize=8)

    # Add relationship annotation
    ax_C1.annotate(f"z‚ÇÇ = -1/zÃÑ‚ÇÅ",
                  xy=(0.05, 0.95), xycoords='axes fraction',
                  color='purple', fontsize=8, va='top')

# Function to visualize orthogonal planes in 4D (represented in 3D)
def visualize_4D_planes(ax, z1, z2, z1_antipodal, z2_antipodal, rotated_point, antipodal_point):
    """
    Visualize the orthogonal C1 and C2 planes in a 3D representation.
    """
    ax.clear()
    ax.set_title("Orthogonal Complex Planes in 4D")

    # Set equal aspect ratio
    ax.set_box_aspect([1, 1, 1])

    # Define the planes
    plane_size = 2.0
    xx, yy = np.meshgrid(np.linspace(-plane_size, plane_size, 10),
                         np.linspace(-plane_size, plane_size, 10))

    # C1 plane (x-y plane at z=0, w=0)
    z_C1 = np.zeros_like(xx)
    ax.plot_surface(xx, yy, z_C1, color='blue', alpha=0.2)

    # C2 plane (z-w plane at x=0, y=0) - we'll represent w as the z-axis and z as the y-axis
    # This is a visual representation of orthogonality in 3D
    ax.plot_surface(np.zeros_like(xx), xx, yy, color='green', alpha=0.2)

    # Draw coordinate axes
    ax.plot([-plane_size, plane_size], [0, 0], [0, 0], 'r-', linewidth=1)  # x-axis
    ax.plot([0, 0], [-plane_size, plane_size], [0, 0], 'g-', linewidth=1)  # y-axis
    ax.plot([0, 0], [0, 0], [-plane_size, plane_size], 'b-', linewidth=1)  # z-axis

    # Label the axes
    ax.text(plane_size+0.1, 0, 0, "Re(C‚ÇÅ)", color='red', fontsize=8)
    ax.text(0, plane_size+0.1, 0, "Im(C‚ÇÅ)", color='green', fontsize=8)
    ax.text(0, 0, plane_size+0.1, "C‚ÇÇ plane", color='blue', fontsize=8)

    # Plot points on C1 plane
    ax.scatter(z1.real, z1.imag, 0, color='blue', s=50, label='z‚ÇÅ')
    ax.scatter(z1_antipodal.real, z1_antipodal.imag, 0, color='cyan', s=30, label='z‚ÇÅ antipode')

    # Plot points on C2 plane (represented on the y-z plane)
    # We're using a visual convention here - the C2 plane is actually orthogonal in 4D
    ax.scatter(0, z2.real, z2.imag, color='green', s=50, label='z‚ÇÇ')
    ax.scatter(0, z2_antipodal.real, z2_antipodal.imag, color='lime', s=30, label='z‚ÇÇ antipode')

    # Draw lines connecting the points to their projections on the planes
    # Line from origin to z1
    ax.plot([0, z1.real], [0, z1.imag], [0, 0], 'b-', alpha=0.7)

    # Line from origin to z2
    ax.plot([0, 0], [0, z2.real], [0, z2.imag], 'g-', alpha=0.7)

    # Draw the projection lines from the sphere points
    # These are visual aids to show how the sphere points project to the planes
    sphere_scale = 0.5  # Scale factor for the sphere representation

    # Scaled sphere points
    sphere_point = rotated_point * sphere_scale
    sphere_antipode = antipodal_point * sphere_scale

    # Add a small sphere to represent the Riemann sphere
    u_sphere = np.linspace(0, 2 * np.pi, 20)
    v_sphere = np.linspace(0, np.pi, 10)
    x_sphere = sphere_scale * np.outer(np.cos(u_sphere), np.sin(v_sphere))
    y_sphere = sphere_scale * np.outer(np.sin(u_sphere), np.sin(v_sphere))
    z_sphere = sphere_scale * np.outer(np.ones_like(u_sphere), np.cos(v_sphere))

    ax.plot_surface(x_sphere, y_sphere, z_sphere, color='lightgray', alpha=0.2)

    # Plot the sphere points
    ax.scatter(sphere_point[0], sphere_point[1], sphere_point[2],
               color='red', s=50, label='Sphere point')
    ax.scatter(sphere_antipode[0], sphere_antipode[1], sphere_antipode[2],
               color='orange', s=50, label='Sphere antipode')

    # Connect sphere points to their projections
    # From sphere point to z1
    ax.plot([sphere_point[0], z1.real],
            [sphere_point[1], z1.imag],
            [sphere_point[2], 0], 'r--', alpha=0.5)

    # From sphere point to z2
    ax.plot([sphere_point[0], 0],
            [sphere_point[1], z2.real],
            [sphere_point[2], z2.imag], 'r--', alpha=0.5)

    # From sphere antipode to z1_antipodal
    ax.plot([sphere_antipode[0], z1_antipodal.real],
            [sphere_antipode[1], z1_antipodal.imag],
            [sphere_antipode[2], 0], 'orange', linestyle='--', alpha=0.5)

    # From sphere antipode to z2_antipodal
    ax.plot([sphere_antipode[0], 0],
            [sphere_antipode[1], z2_antipodal.real],
            [sphere_antipode[2], z2_antipodal.imag], 'orange', linestyle='--', alpha=0.5)

    # Add text showing the antipodal relationship
    ax.text(0.5, 0.05, 0.95, f"z‚ÇÇ = -1/zÃÑ‚ÇÅ", transform=ax.transAxes,
            fontsize=10, color='purple')

    # Set view angle
    ax.view_init(elev=30, azim=45)

    # Set limits
    ax.set_xlim(-plane_size, plane_size)
    ax.set_ylim(-plane_size, plane_size)
    ax.set_zlim(-plane_size, plane_size)

    # Add legend
    ax.legend(loc='upper right', fontsize=8)

# Animation
def update(frame):
    ax_sphere.clear()
    ax_C1.clear()
    ax_C2.clear()

    # Rotate the SPHERE (not the point)
    theta = frame * 0.05  # Rotation speed

    if frame % 2 == 0:  # North-South (y-axis rotation)
        rot_matrix = np.array([
            [np.cos(theta), 0, np.sin(theta)],
            [0, 1, 0],
            [-np.sin(theta), 0, np.cos(theta)]
        ])
        rotated_x = rot_matrix[0, 0] * x + rot_matrix[0, 2] * z
        rotated_z = rot_matrix[2, 0] * x + rot_matrix[2, 2] * z
        rotated_point = rot_matrix @ fixed_point
        ax_sphere.set_title("N-S Rotation (Y-axis)")
    else:  # East-West (x-axis rotation)
        rot_matrix = np.array([
            [1, 0, 0],
            [0, np.cos(theta), -np.sin(theta)],
            [0, np.sin(theta), np.cos(theta)]
        ])
        rotated_y = rot_matrix[1, 1] * y + rot_matrix[1, 2] * z
        rotated_z = rot_matrix[2, 1] * y + rot_matrix[2, 2] * z
        rotated_point = rot_matrix @ fixed_point
        ax_sphere.set_title("E-W Rotation (X-axis)")

    ax_sphere.plot_surface(rotated_x if 'rotated_x' in locals() else x,
                           rotated_y if 'rotated_y' in locals() else y,
                           rotated_z if 'rotated_z' in locals() else z,
                           color='lightblue', alpha=0.2)

    # Plot fixed point (rotated with sphere)
    ax_sphere.scatter(*rotated_point, color='red', s=50, label='Point')

    # Add the antipodal point on the sphere
    antipodal_point = -rotated_point  # Negate all coordinates
    ax_sphere.scatter(*antipodal_point, color='orange', s=50, label='Antipode')

    # Draw a line connecting the points on the sphere
    ax_sphere.plot([rotated_point[0], antipodal_point[0]],
                  [rotated_point[1], antipodal_point[1]],
                  [rotated_point[2], antipodal_point[2]],
                  'k--', alpha=0.5)

    ax_sphere.set_xlim(-1, 1); ax_sphere.set_ylim(-1, 1); ax_sphere.set_zlim(-1, 1)
    ax_sphere.legend(loc='upper right', fontsize=8)

    # Project both points to both planes
    z1_direct = project_C1(rotated_point)
    z2_direct = project_C2(rotated_point)

    z1_antipodal = project_C1(antipodal_point)
    z2_antipodal = project_C2(antipodal_point)

    # Plot ‚ÑÇ1 (stereographic projection)
    ax_C1.scatter(z1_direct.real, z1_direct.imag, color='blue', s=50, label='Direct')
    ax_C1.scatter(z1_antipodal.real, z1_antipodal.imag, color='cyan', s=30, label='Antipode')
    ax_C1.add_artist(plt.Circle((0,0), 1, fill=False, linestyle='--', color='red'))
    ax_C1.set_xlim(-3, 3); ax_C1.set_ylim(-3, 3)
    ax_C1.set_title("‚ÑÇ1: Stereographic")
    ax_C1.legend(loc='upper right', fontsize=8)

    # Plot ‚ÑÇ2 (antipodal projection)
    ax_C2.scatter(z2_direct.real, z2_direct.imag, color='green', s=50, label='Direct')
    ax_C2.scatter(z2_antipodal.real, z2_antipodal.imag, color='lime', s=30, label='Antipode')
    ax_C2.add_artist(plt.Circle((0,0), 1, fill=False, linestyle='--', color='red'))
    ax_C2.set_xlim(-3, 3); ax_C2.set_ylim(-3, 3)
    ax_C2.set_title("‚ÑÇ2: Antipodal")
    ax_C2.legend(loc='upper right', fontsize=8)

    # Add relationship visualization
    add_relationship_visualization(ax_C1, ax_C2, z1_direct, z2_direct, z1_antipodal, z2_antipodal)

    # Visualize the orthogonal planes in 4D (represented in 3D)
    visualize_4D_planes(ax_4D, z1_direct, z2_direct, z1_antipodal, z2_antipodal,
                       rotated_point, antipodal_point)

    plt.tight_layout()
    return ax_sphere, ax_C1, ax_C2, ax_4D

# Run animation
'''ani = FuncAnimation(fig, update, frames=120, interval=50, blit=False)
plt.close()
HTML(ani.to_jshtml())'''
```

### Proof that XOR-XNOR logic form logical duality (with a surprise)

**1. Truth Table ‚Üí 4D Vectors (No Antipodality Yet)**

The **XOR** and **XNOR** truth tables for inputs $(A, B) \in \{0,1\}^2$ yield 4D outputs when fully enumerated:  
‚Ä¢ **XOR outputs**: $(0, 1, 1, 0) (for inputs (0,0), (0,1), (1,0), (1,1))$.  
‚Ä¢ **XNOR outputs**: $(1, 0, 0, 1)$ (logical complement of XOR).


**Observation**: These vectors are **orthogonal** but **not antipodal**:  
They (add to (1,1,1,1)) and
  
$(0,1,1,0) \cdot (1,0,0,1) = 0 + 0 + 0 + 0 = 0$.
    
**Surprise**: They *are* orthogonal! But this is **not sufficient** to span orthogonal planes.  


‚∏ª


**2. From Vectors to Orthogonal Planes**

For **two orthogonal planes** in ‚Ñù‚Å¥, we need:  
‚Ä¢ **Plane 1**: Spanned by two linearly independent vectors (e.g., $\mathbf{u}_1 = (0,1,1,0), \mathbf{u}_2 = (1,0,0,1)$).  
‚Ä¢ **Plane 2**: Spanned by two vectors **orthogonal to all vectors in Plane 1**.


**Problem**: The XNOR vector (1,0,0,1) is **already in Plane 1** (as $\mathbf{u}_2$). We need **new orthogonal vectors** for Plane 2.  


**Constructing Plane 2**:
‚Ä¢ Find vectors $\mathbf{v}_1, \mathbf{v}_2$ such that:  
  
$\mathbf{v}_i \cdot \mathbf{u}_j = 0 \quad \forall i,j \in \{1,2\}$.
    
‚Ä¢ Solutions (via Gram-Schmidt):  
  
$\mathbf{v}_1 = (0, 1, -1, 0), \quad \mathbf{v}_2 = (1, 0, 0, -1)$.
    
‚Ä¢ **Verification**:  
  
$(0,1,1,0) \cdot (0,1,-1,0) = 0 + 1 -1 + 0 = 0 \quad \checkmark$
    

$(1,0,0,1) \cdot (1,0,0,-1) = 1 + 0 + 0 -1 = 0 \quad \checkmark$



Now we have:  
‚Ä¢ **Plane 1 (XOR-XNOR-like)**: $\text{span}\{(0,1,1,0), (1,0,0,1)\}$.  
‚Ä¢ **Plane 2 (signed XOR-XNOR-like)**: $\text{span}\{(0,1,-1,0), (1,0,0,-1)\}$.


‚∏ª

**3. Rotating to Canonical Form $(x_1,x_2,0,0) \ and \ (0,0,x_3,x_4)$**

We want a rotation matrix **Q** such that:  
  
$Q \text{ maps Plane 1} \to (x_1, x_2, 0, 0), \quad Q \text{ maps Plane 2} \to (0, 0, x_3, x_4)$.
  


**4. Build \(Q\) from Orthonormalized Bases**
1. Normalize vectors (call them $\mathbf{u}_1, \mathbf{u}_2$ and $\mathbf{v}_1, \mathbf{v}_2$).
2. **Q** is the matrix with rows = these orthonormal vectors:  
  
$Q = \begin{bmatrix}
0 & \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} & 0 \\
\frac{1}{\sqrt{2}} & 0 & 0 & \frac{1}{\sqrt{2}} \\
0 & \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} & 0 \\
\frac{1}{\sqrt{2}} & 0 & 0 & -\frac{1}{\sqrt{2}}
\end{bmatrix}$.
    
3. **Apply Q**:  
  
$Q \mathbf{u}_1 = (1, 0, 0, 0), \quad Q \mathbf{u}_2 = (0, 1, 0, 0)$,
    

$Q \mathbf{v}_1 = (0, 0, 1, 0), \quad Q \mathbf{v}_2 = (0, 0, 0, 1)$.



**Result**:  
‚Ä¢ Plane 1 becomes $(x_1, x_2, 0, 0)$.  
‚Ä¢ Plane 2 becomes $(0, 0, x_3, x_4)$.


‚∏ª

The **Q matrix**,

$Q = \begin{bmatrix}
0 & \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} & 0 \\
\frac{1}{\sqrt{2}} & 0 & 0 & \frac{1}{\sqrt{2}} \\

\end{bmatrix}$.

is the matrix with the **XOR-XNOR** plane and its signed version and their respective vectors set in order, multiplied by the coefficient 1/‚àö2.

‚∏ª


**5. Final Picture: Antipodality on $S^2$**
1. **Orthogonal planes** ‚Üí Rotate to $(x_1, x_2, 0, 0) and (0, 0, x_3, x_4)$.  
2. **Map to ‚ÑÇ¬≤**:  
  
$(x_1, x_2, x_3, x_4) \mapsto (z_1 = x_1 + i x_2, z_2 = x_3 + i x_4)$.
    
3. **Stereographic projection**:

- signed-XOR-XNOR plane $(z_1 = 0) ‚Üí \ 0 \ on \ S^2$.


**Antipodality**: \(0\) and $\infty$ are antipodal on the Riemann sphere.

**6. Recursive Symmetry via M√∂bius Maps**
‚Ä¢ M√∂bius transformations preserve orthogonality and antipodality.  
‚Ä¢ The map $f(\zeta) = \frac{1}{\zeta}$ swaps:
