## Further developments – Proposed Framework towards Personalised Medicine

So far, we have developed the LEON core algorithm and demonstrated that using an ensemble method with the core algorithm (that uses logical recursion modulated by Dirichlet-Multinomial distribution) permits us to engage in feature extraction of simple images (digits) and Chromosomal analysis with a high degree of accuracy. This hybrid method is expected to provide highly accurate feaature extraction and prediction. Thus it is postulated that this could be applied in other contexts such as more complex imaging (fMRI) or time series data (EEG). While these conclusions should be approached cautiously and further testing and development of the code is required, it appears a worthwhile exercise. Further, strengthening this model of logical recursion and concentration of probabilities on a multinomial tree is the independent capacity to simulate plausible biological feature of membrane spikes. We can also link this finding to  the capacity of the LEON model to find features as attractors in the state space represented by plausible empirical biological processes. Importantly, the model used available empirical data and models to derive a deductive model that is canonical and apllicable across all scales and levels in the Biopsychosocial framework. We have demonstrated the plausibility of using LEON to model and predict in geneomics and membrane function. Below, we explore the possibility of testing this model further by applying a concurrent validation programme for all biological levels (at present confined to brain function). In this regard we explore possible investigative methods from single neurons to Large Scale Networks.

### Single Neuron simulations – Comparison with NEURON

Single Neuron simulations have already been carried out using NEURON simulator as the gold standard in which a neural network model has been validated. It is possible to develop neural network models that are congruent with specific pattern of activity in a single neuron including type of input, logic operations and level of complexity of information processing even in a single neuron (a study has demonstrated that a single neuron could be the equivalent of up to a 7 level deep neural network. This would be congruent with we have postulated in the model presented here.

Analogously, it is our intention to test the LEON algorithm as a core such that a an atrifical neural network developed by it is congruent with the activity of a single neuron. We expect that the features selected by the algorithm (Dirichlet Multinomial) would be attractors of the recursion process, that could be aligned to possible biological processes (these could be input features, in a manner similar to how we have postulated that the factors that influence the simulation of the membrane spikes could be asociated with diferent expression features of ion channels)

We postulate that applying the LEON algorithm to levels of brain function, from cortical modules/columns to Large Scale Networks (even Left Right hemisphere differences) would be associated with DM-derived features that could be interpreted as input or interactive factors in these systems. **Interpreting the Channel States**

Firstly let us consider the regions covered by the channels above:
  
**Fp1:** Left Prefrontal Region
  
**F3:** Left Frontal region

**F4:** Right Frontal Region

**T4:** Right Mid-Temporal Region

**T5:** Left Posterior-Temporal Region

**Pz:** Midline Parietal Region

**C3:** Left Central Region (Central Sulcus - Somato Sensory)

**O2:** Right Occipital Lobe (including projections to Cuneus)

Examining the table above, and focussing on the strong suppression of **fractal dimension** following treatment we can see that frontal activity is reduced to a fractal dimension of 4-5 while the right mid temporal (during photic stimulation) and left posterior temporal regions (when eyes open) gave suppressed fractal dimensions. Photic stimulation was also associated with a reduction in the left somatosensory and posterior temporal region.

The strong facilitation of fractal dimension are expressed in the midline parietal (possibly Precuneus?) in the eyes closed state and left frontal, right mid temporal and right occipital in the photic stimulation conditiopn. The left frontal lobe demonstrates increased activity in both states after treatment.

If we consider these findings in terms of large scale networks we can infer that the treatment has an impact in the following ways:
Any excessive frontal activity (excess input from too many systems) is reduced but prefrontal activity (in the left is increased). Interestingly even with the eyes open and during stimulation there is no excess somatosensory reaction and the path conveying sensory information to the hippoicampus may be modulated.
  
How this is achieved may be gleaned from looking at the active regions. The **left prefrontal cortex** appear to have increased activity (increased **number of systemic inputs**) whether inwardly focussed (eyes-closed) or stimulated state, indicating it is consistently in control. Importantly, when in the **eyes-closed state** the midline parietal region becomes active and we could infer that ego-centric body awareness is increased (less distractibility), possibly via precuneus and posterior cingulate gyrus interaction. Left frontal region becomes active during stimulatory condition (perhaps modulated by the prefrontal input) but interestingly, it is the **contralateral temporal lobe** that is concurrently active (as well as the right occipital pole). When in a treated state, this individual's brain is better able to be coordinated in which it responds to **contemplation** versus **stimulated state**. When stimulated, the brain switches to **right hemispheric processing**, indicting a search paradigm for novel non-threatening cosntructs. In the treated state then we could postulate that the **DMN** is able to switch more smoothly from contemplative inwardly focussed states to sensorimotor states (**CEN and SN**) smoothly. 
  
It is also likely that **excess input** to **frontal structures** are suppressed (perhaps by inhibitory influence of the **prefrontal** region. The observation that treatment actually reduces temporal response to **epileptogenic stimuli** and in eyes open condition but when stimulation is sustained the left temporal function continues to be reduced while increasing the activity of the **right mid-temproal** region.

We have used words like increased activity here. However, what we are really seeing is possibly increased or decreased independent inputs to these regions. This is discussed further below.

### What are the input systems that alter the Correlation Dimension of the EEG

We consider the following inputs on the background of thalamo-cortical relay that contribute to the variability of the EEG:
  
Thalamocortical Glutamate (Rhythmic Pacing)
  
Intra-module Recurrent Glutamate (Signal Amplification)
  
Transcortical GABA (Inhibitory Gain Control): bring the system back to equilibrium (bit-flipping)
  
Serotonergic (5-HT) (Excitability Tuning): Novelty seeking
  
Adrenergic/Dopaminergic: (NA/DA) (Signal-to-Noise Modulation): HCN channel modulation at the early phase (see below)
  
Cholinergic (ACh) (Arousal/Attention Gating): minimising novelty seeking

**Broad Functional Differences between 5HT, NA and DA**

Both serotonin and noradrenalin systems can be dmonstrated in lesion and stimulation studies to respond to novel stimuli. Their response set differ and could be considered complementary:

5HT - responds to novel non-threatening stimuli: broad function is to induce a curious-exploratory state when optimnally tuned.In animal studies (including ethology), it is possible to show that optimal serotonin function contributes to dominance hierarchy. The alpha male generally has evidence of optimal serotonin function. However, if thier serotonin system is lesioned, then they drop down the hierarchy. The males in the lower rungs of the hierarchy tend to be generally more aggressive and reactive and hypervigilant. It appears that the readiness to respond to novel stimuli has to be balanced by capacity remain calm against threat.

NA- responds to novel or known threatening stimuli. While both 5HT and NA have positive feedback on each other during optimum circumstances, this balance can shift if the threat levels escalate. In such circumstance, serotonin function is inhibited and NA function escalates. NA is cruciall in early information processing (50 ms - P50) at the hippocampus.

Monoamines project to all brain regions including frontal regions where motor decision are formulated and transmitted to cortical and subcortical motor centres. It could be inferred that under optimum conditions, the balance between serotonin, noradrenalin and dopamine are finely balanced at the frontal lobes and most decisions even when they are habits can have conscious control. (The frontal region comes into play at about 2-300 ms - P300).

We postulate, when, due to an intense signal, NA activity escalates, and 5HT system is inhibited, the connection between the Large Scale Networks is broken and the evolutionarily earlier (lower) circuits take over. Typically, this would involve the approach-avoidance circuits from the amygdala to the ventral striatum. Once a habit is formed in this circuit, it acts like a limit cycle as opposed to a complex fractal pattern. The attractor is not complex.

Dopamine is the system that targets mass motor actions and is a signifcant motivational system. When the system swiitches from a frontally controlled to a subcortical sensori-motor-behavioural pattern, then there is already an established interaction by the Dopamine system from the ventral tegmentum to the Nucleus Accumbens (Ventral Striatum).
  
This interaction between the three system is probably generalised along cortical modules and individual neurons. As an example of the complementary interaction of these systems is their interaction at HCN channels on the dendritic arbour. The most likely interaction is the activation HCN channels via beta-2 adrenergic receptors, while 5HT and DA generally inhibit HCN channel activity via their recptors. There are some 5HT receptors that may have an opposite effect.

While, the interaction between these systems are probably more complex than described here, this model provides reasons to acknowledge that aminergic input in the cortex may signifcantly affect the EEG output due to their modulation of the underlying Glutamate induced input and amplification at the cortical modules. However, due to specific effects on particualr cortical area, the impact of aminergic input may have different effects, resulting in the sort of pattern described above.


**Connecting the dots with fMRI and social navigation**

In the last two decades, significant advances have been made in the understanding of social navigation and how humans classify hierarchy and affiliation and their place in such an abstract landscape. We have previously discussed a plausible hypothesis for the evolution of the structures of social navigation from more primitive structures for phsysical navigation. We established a model of antipodal relations between pairs of th 12 basic emotions that could produce complex emotional experience as land marks (and landscape) for making judgements about behavioural plans. These anrtipodal relations and rotations were linked to feed-forward feed back relations between cortical modules and possibly between the Large Scale Networks(LSN). Here we attempt a cuatious model of how they relate to each other and how monoamine projections could influence such interaction (and be a proxy for measuring how the large Scale Networks are functioning).

Here we attempt to build an explanatory model how the higher LSN's interact with each other and lower circuits and how aminergic input to these striucture may help modulate the functional relationships broadly. The connections and their function act as the recurrent features (XOR-XNOR recurrence) of which the the recurrent connection between the Left (XOR) and Right (XNOR) hemispheres may be the ultimate complex sytem in which the logical complements play out. The aminergic projections could be seen as the modulatory inputs that influence the probability distribution of firing of nodes in the system, whether at a single neuron levels or LSN level. As we have seen, there are reasons to think that at the membrane level and gene expression level, different constraints act to limit the expression probabilities.

Below, we postulate how the three LSN may interact with each other based on known evidence. DMN, in this putative model is consdered to sitch the networks from equilibrium (DMN internally focussed) to externally focussed in which its activity diminishes and the recurrent activity between SN and CEN operates to bring a new equilibrium following new information. The switching by DMN could represent antipodal switching of emotional salience and the recurrence between the others would be an exploration of the new landscape, to establish an attractor state.

In social navigation, fMRI studies have demonstrated that social distance is estimated by the Default Mode Network and posterior cingulate cortex and studies that have uncovered two dimensional social space appear to find an interaction between these structures including the percuneus and the hippocampus in tracking one's postion in the social landscape. It is intersting that the psoterior aspects of the DMN is consistently implicated in social distance estimates that appears to involve vector angles (much like in landmarks). These findings support the idea of the DMN contributing to emotions and social navigation by providing virtual (antipodal) rotations. However, as these processes appear to merely involve recurrent circuits, this meets our criteria for XOR-XNOR recursion. As indicated previously, whilst this is a discrete process, the sheer number of nodes created among thousands of interactions per cycle with an enormous state space (e.g. 10^{300}) would make the process appear continuous.

**How do the aminergic projection affect the LSN's**

As discussed above, we propose that the DMN acts as a switching circuit for the three networks to switch from equilibrium to search functions. It has been demonstrated that plausibility of XOR-XNOR recursion and their relationship to continuous function can be established using a mutinomial tree and recursion. Here we propose further interaction by aminergic projctions that affect the broad function of Large Scale Netowrks and Hemisphere function by "potential-differential" impact they have on different nodes of these large structures.

As inidcated previo,usly the 5HT and NA respond to novel stimuli but differently. DA will generally respond to the state of balance between these two systems by facilitating mass motor actions (via striatal connections). However, the interaction between these systems and with the large networks and systems is not likley to be that straight forward. This can be gleaned by considering the number of different types of receptors, particulalry in the seorotonin (5HT) system.

In the example of the EEG fractal dimension map above, we saw how particular cortical area required either lowering of dimension or increasing the dimension with treatment. The recurrent relationships between the aminergic systems needs to be established but likely to be very complex, especially when taking into account their relationship with descending inputs from higher structures. These projections have evolved for hundreds of million years and developed a wide range of network connections. However, for reasons of parsomony, it is always useful to go back to thier broad functions. As outlined above, we postulate that there are switches that occur in the balance between NA and 5HT dependng on context. When the system is pushed into imbalance, it can descend to use more primitive emotions and circuits from which it is diffiuclt find a path back to previous equlibrium. Because these are evolutionarily older circuits, they may be particularly resistant to shifting without understanding how to shift them (psychotherapy or medications). In these circuits (e.g. Amygdala and Ventral striatum associated with the hippocampal context provision) emotions guided by construct such as fight or flight state (the antipodal feelings being anger and fear) operate to produce navigation in the environment. The Amydala-Ventral Striatal ciircuit probably serves as a "rotator axis" in this situation. However, these circuits are also signifcantly controlled by both descending inputs (cortical striatal loops) and ascending circuits from the RAS (the aminergic projections). If the serotonin function has been overcome by excess activity of the NA system, this dissociates this lower circuit from the Large Cortical Networks. Such stuck states can only be overcome by breaking and modulating the influence of aminergic projections and engaging in therapy to re-establish the large Scale Networks.

Always, in these interventions we expect to shift from one attractor (equlibrium) to another. Such attractors are complex-chaotic in nature and can be measured via the fractal dimension. We assert here that the recurrent logical interactions mediate these changes and occur in a scale independent manner. Thus by further investigation and training, it may be possible to develop an AI system that can flexibly recognise these patterns and make predictions and engage in modellling. We consider such applications briefly below. 

### Applications:

Using the same canonical coupled algorithm it is proposed that multiple scales of brain function can be analysed with potential for predictive and simulation based modelling:

Once the concurrent validity of an EEG mapping against fMRI is performed, the former can be related to BOLD signal outcomes. These outputs can be used to further infer emotional spectrum of an individual and predict outcomes of interventions using machine learning algorithms. (This study demonstrates that there is potential for identifying important biological parameters in terms of empirically derived measures at any level of the biopsychosocial spectrum; in the case of the EEG, if it is correlated in terms of fractal dimension of cortical areas to fMRI of social navigation, then, the fractal dimension could be modelled using the algorithms developed (LEON and ensemble models) to reverse engineer and predict balance of activity between typical brain systems that contribute to the EEG via their mass actions (e.g. aminergic projections). Thus the AI algorithm could be used iteratively with a EEG model of social navigation to infer likely activity level of important pharmacodynamic systems in individual patients.)

3)	If coupled with peripheral genomic analysis of immune cells, this may provide inference about the longitudinal impact on aminergic systems

5)	This iterative learning process would be ideal for generating plausible models of brain network functions as well as finding markers of longitudinal impact on brain function as well as systemic impact (including the social system)
