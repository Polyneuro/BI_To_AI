### Final Möbius Membrane Dynamics-Sphere Rotation

In this final experiment, we have provided aan extra event cluster and and alpha range that is narrower 0.7-1.1. The step sizes were reduced to 6. This appears to produce the expected spike train that is consistent.

```python
# Sphere rotation block. import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
# Add these at the top of your code (before creating the figure to avoid animation size limit)
#plt.rcParams['animation.embed_limit'] = 50  # Increase to 30MB (default is 20MB)
#plt.rcParams['figure.dpi'] = 72  # Reduce from default 100 dpi
# Animation settings
plt.rcParams['animation.embed_limit'] = 100 # Increased from 45 to 100 MB
plt.rcParams['figure.dpi'] = 72

# Setup figure
fig = plt.figure(figsize=(16, 6))
ax_sphere = fig.add_subplot(131, projection='3d')
ax_membrane = fig.add_subplot(132)
ax_conductance = fig.add_subplot(133)

# Parameters
V_rest = -70.0
V = V_rest
spike_threshold = -55
reset_potential = -80
total_frames = 300
tau_m = 7.0
refractory_period = 2

# Enhanced synaptic parameters with DM-distributed inputs
event_clusters = {
    'early': np.arange(30, 110, 6),  # event clusters between 30-110 frames , step size =6)
    'late': np.arange(140, 210, 6),
    'extra': np.arange(220, 290, 6)
}

# Optimized DM probability parameters
alpha = [0.9, 0.7, 1.1]
synaptic_base_weight = 14
synaptic_decay = 0.88
summation_buffer = []

# Reversal potentials and conductances
E_Na = 60.0
E_K = -90.0
E_Cl = -65.0
g_Na_max = 1.2
g_K_max = 0.7 # Increased from 0.5 for stronger K+ effect
g_Cl_max = 0.1

# Initialize Riemann sphere
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x_sphere, y_sphere, z_sphere = np.cos(u)*np.sin(v), np.sin(u)*np.sin(v), np.cos(v)
ax_sphere.plot_wireframe(x_sphere, y_sphere, z_sphere, color="gray", alpha=0.2)

# Initialize plots and state
time_points = []        # The sphere point is used as time points on the plot
voltage_history = []    # voltage_hitory has voltage appended and is plotted against time points
I_Na_hist = []          # all gate currents and voltages are calculated according to sphere rotation
I_K_hist = []
I_Cl_hist = []
g_Na_hist = []
g_K_hist = []
g_Cl_hist = []
synaptic_current_hist = []
point = np.array([0, 0, -1])    # starting point on X,Y,Z plot on sphere
refractory_counter = 0
rotation_axis = 0
rotation_noise = 0.08

# For K+ conductance delay - longer buffer
k_activation_buffer = [0] * 4 # 4-frame buffer (was 3)
prev_g_Cl = g_Cl_max * 0.2

# Plot objects
point_plot = ax_sphere.scatter([], [], [], color='red', s=100)
voltage_line, = ax_membrane.plot([], [], 'k-', lw=2, label="Membrane Potential")
synaptic_line, = ax_membrane.plot([], [], 'b:', alpha=0.5, label="Synaptic Inputs")
threshold_line = ax_membrane.axhline(spike_threshold, linestyle='--', color='r', alpha=0.3)

def get_conductances(point, V):
    global k_activation_buffer, prev_g_Cl

    # Enhanced voltage gating with smoother activation
    v_gate = 1 / (1 + np.exp(-2.5*(V - (V_rest+spike_threshold)/2)/(spike_threshold-V_rest))) #v_gate ~ V related exponential

    # Use z-coordinate to control conductances
    z = point[2]

    # Na+ active at North Pole (z > 0) with sharper activation
    g_Na = g_Na_max * (1 + z)/2 * v_gate**1.5

    # K+ active at South Pole (z < 0) with modified activation
    raw_g_K = g_K_max * (1 - z)/2 * v_gate**2.0 # Reduced from 2.5 to 2.0

    # Implement K+ activation delay using longer buffer
    k_activation_buffer.append(raw_g_K)
    delayed_g_K = k_activation_buffer.pop(0)

    # Amplify K+ conductance during repolarization
    if V < 0 and V > -40: # During repolarization phase
        delayed_g_K *= 1.5 # Boost K+ conductance

    # Cl- as a smoothed leak conductance with reduced magnitude
    raw_g_Cl = g_Cl_max * (0.2 + 0.8*(1 - v_gate))
    g_Cl = prev_g_Cl * 0.8 + 0.2 * raw_g_Cl
    prev_g_Cl = g_Cl

    return g_Na, delayed_g_K, g_Cl

def update(frame):
    global V, point, summation_buffer, refractory_counter, rotation_axis

    # DM-distributed synaptic inputs with optimized path convergence
    if any(frame in cluster for cluster in event_clusters.values()) and refractory_counter == 0:
        # Generate DM-distributed weights with higher probability of strong paths
        weights = np.random.dirichlet(alpha)

        # Bias toward stronger weights (path convergence)
        if weights[0] > 0.5: # If first component is dominant
            weights[0] *= 1.2 # Boost it further
            weights = weights / np.sum(weights) # Renormalize

        effective_weight = synaptic_base_weight * (1 + weights[0] - 0.7*weights[1]) # Reduced inhibition

        summation_buffer.append({
            'amp': effective_weight,
            'decay': synaptic_decay * (0.9 + 0.2*weights[2]) # Less decay variability
        })

    # Enhanced temporal summation with non-linear integration
    Isyn = 0
    active_synapses = len(summation_buffer)

    # Non-linear summation boost with stronger effect
    if active_synapses > 0:
        boost_factor = 1 + 0.6 * np.log1p(active_synapses**0.9) # Stronger boost
        for syn in summation_buffer[:]:
            Isyn += syn['amp'] * boost_factor
            syn['amp'] *= syn['decay']
            if syn['amp'] < 0.5:
                summation_buffer.remove(syn)

    # Membrane dynamics
    if refractory_counter > 0:
        refractory_counter -= 1
        if refractory_counter == 0:
            V = reset_potential + np.random.randn() # Less noise in reset
            rotation_axis = 0
    else:
        # Voltage integration with enhanced near-threshold dynamics
        deltaV = (V_rest - V + Isyn) * (1 - np.exp(-1/tau_m))      # Voltage follows HH dynamics for summation

        # Enhanced subthreshold boost with smoother transitions
        if V > -58:
            subthreshold_boost = 2.0 + 0.5 * np.tanh((V + 55)/2) # Smooth boost near threshold
        elif V > -65:
            subthreshold_boost = 1.2
        else:
            subthreshold_boost = 1.0

        V += deltaV * subthreshold_boost + 0.8 * np.random.randn() # Reduced noise

    # Spike generation
    if V >= spike_threshold and refractory_counter == 0:
        refractory_counter = refractory_period
        summation_buffer = []       # stochastic- determined by DM
        point = np.array([0, 0, 1]) # North pole (Na+ dominance)
        rotation_axis = 1
        V = 35.0 + 3*np.random.randn() # Higher peak with less variability

    # Rotation dynamics with improved stability
    if refractory_counter == 0:
        # Voltage-dependent rotation speed
        theta = 0.05 * (1 + 1.2*(V - V_rest)/(spike_threshold-V_rest))
        theta = np.clip(theta, 0.01, 0.2) # Limit rotation speed

        # Add controlled noise to rotation axis
        axis_noise = rotation_noise * np.random.randn(3)
        if rotation_axis == 0: # ℂ1 (K+)
            axis = np.array([1, 0, 0]) + axis_noise
        else: # ℂ2 (Na+)
            axis = np.array([0, 1, 0]) + axis_noise
        axis /= np.linalg.norm(axis)

        # Rodrigues' rotation formula
        rot_matrix = np.cos(theta)*np.eye(3) + \
                     np.sin(theta)*np.array([[0, -axis[2], axis[1]],\
                                             [axis[2], 0, -axis[0]],\
                                             [-axis[1], axis[0], 0]]) + \
                     (1-np.cos(theta))*np.outer(axis, axis)

        point = rot_matrix @ point       # matrix multiplication (@) of point vector by rotation matrix

        # Improved context-dependent switching with voltage influence
        switch_prob = 0.02 * (1 + Isyn/15 + np.abs(V - V_rest)/20)
        if np.random.rand() < switch_prob:
            rotation_axis = 1 - rotation_axis

        # NEW: Enhance K+ activation after Na+ phase
        if rotation_axis == 0 and V > -60: # Switching to K+ mode after depolarization
            # Temporarily boost K+ by moving point toward South pole
            point = point * 0.8 - np.array([0, 0, 0.2])
            point /= np.linalg.norm(point) # Renormalize

    # Get conductances
    g_Na, g_K, g_Cl = get_conductances(point, V)

    # Store data
    time_points.append(frame * 0.1)
    voltage_history.append(V)
    synaptic_current_hist.append(Isyn)

    # Store conductances and currents
    g_Na_hist.append(g_Na)
    g_K_hist.append(g_K)
    g_Cl_hist.append(g_Cl)

    I_Na_hist.append(g_Na * (V - E_Na))
    I_K_hist.append(g_K * (V - E_K))
    I_Cl_hist.append(g_Cl * (V - E_Cl))

    # Update plots
    point_plot._offsets3d = ([point[0]], [point[1]], [point[2]])
    voltage_line.set_data(time_points, voltage_history)
    synaptic_line.set_data(time_points, synaptic_current_hist)

    # Conductance plot with adjusted K+ scaling for better visibility
    ax_conductance.clear()
    ax_conductance.plot(time_points, np.array(g_Na_hist)/g_Na_max*0.7, 'b-', label="Na+ conductance")
    ax_conductance.plot(time_points, np.array(g_K_hist)/(g_K_max*1.2), 'g-', label="K+ conductance") # Scaled for visibility
    ax_conductance.plot(time_points, np.array(g_Cl_hist)/g_Cl_max, 'm-', label="Cl- conductance")
    ax_conductance.legend()
    ax_conductance.set_title("Channel Conductances (Normalized)")

    # Formatting
    ax_membrane.set_xlim(0, max(10, frame*0.1))
    ax_membrane.set_ylim(-90, 40)
    ax_membrane.set_title("Membrane Potential with Summation")
    ax_membrane.legend(loc='upper right')

    sphere_title = "ℂ1 (K+ gating)" if rotation_axis==0 else "ℂ2 (Na+ gating)"
    ax_sphere.set_title(f"Riemann Sphere: {sphere_title}")

    return point_plot, voltage_line, synaptic_line

# Generate animation
#ani = FuncAnimation(fig, update, frames=total_frames, interval=50, blit=False)
#plt.close()
#HTML(ani.to_jshtml())
```
